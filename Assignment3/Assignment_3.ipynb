```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/gauraviiita/Supervised_ML/main/Datasets/Chapter_2/data.csv")

print(data.shape)
print(data.info())

print(data.isna().sum())

mean = data['Calories'].mean()
data['Calories'] = data['Calories'].fillna(mean)
print(data.isna().sum())

data.dropna(inplace=True, axis=0)
print(data.isna().sum())

data['Date'] = pd.to_datetime(data['Date'], format='mixed')
print(data.head())

for x in data.index:
    if(data.loc[x, 'Duration']>60):
        data.loc[x, 'Duration']=60
print(data.head())

print(data.duplicated().sum())
data.drop_duplicates(inplace=True)
print(data.duplicated().sum())

data.to_csv('Cleaned data.csv', index=False)
```

```python
import numpy as np 

def xy_label(x="<--X-axis-->", y="<--Y-axis-->", title=""):
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(title)

arr = np.linspace(-np.pi, np.pi, 200)

sin_arr, cos_arr, tan_arr = np.sin(arr), np.cos(arr), np.tan(arr)

plt.figure(figsize=(8, 10))
plt.subplot(3, 1, 1)
plt.plot(arr, sin_arr)
xy_label(title="Sin Graph")

plt.subplot(3, 1, 2)
plt.plot(arr, cos_arr)
xy_label(title="Cos Graph")

plt.subplot(3, 1, 3)
plt.plot(arr, tan_arr)
xy_label(title="Tan Graph")

plt.tight_layout()
```

```python
import numpy as np
from numpy import linalg

arr_1 = np.array([[1, 2], [3, 4]])
arr_2 = np.array([[2, 4], [6, 9]])

print(np.dot(arr_1, arr_2))
print(linalg.eigh(arr_1))
print(linalg.eigh(arr_2))

print(linalg.det(arr_1))
print(linalg.det(arr_2))

print(linalg.inv(arr_1))
print(linalg.inv(arr_2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def xy_label(x="", y="", title=""):
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(title)

data = pd.read_csv("https://raw.githubusercontent.com/gauraviiita/Supervised_ML/main/Datasets/Chapter_2/ComputerSales.csv")

print(data.info())
j=1
row = 2
column = int(np.ceil(5/row))

print(row, column)

# plt.figure(figsize=(8, 13))
for i in data:
    if(str(data[i].dtypes)=="int64" or str(data[i].dtypes)=="float64" or str(data[i].dtypes)=="int32" or str(data[i].dtypes)=="float32" ):
        plt.subplot(row, column, j)
        plt.hist(data[i], edgecolor="black")
        xy_label(x=i, y="Frequency")
        j+=1
plt.tight_layout()
```

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("https://raw.githubusercontent.com/gauraviiita/Supervised_ML/main/Datasets/Chapter_2/police.csv")
print(data.info())
print(data.isna().sum())

data.dropna(inplace=True, axis=1, how='all')
print()
print(data.isna().sum())
print(data.info())

for i in data:
    if(str(data[i].dtype)=="float64"):
        data[i] = data[i].fillna(data[i].mean())
print()
print(data.isna().sum())
print(data.info()) 

data.dropna(inplace=True, axis=0) 
print()
print(data.isna().sum())
print(data.info())    

print()
print(data.duplicated().sum())
data.drop_duplicates(inplace=True)
print(data.duplicated().sum())
print(data.info())

data['stop_date'] = pd.to_datetime(data['stop_date'], format='mixed')
data['stop_time'] = pd.to_datetime(data['stop_time'], format='mixed').dt.time
print()
print(data.info())

sns.pairplot(data)
```
