```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import svm
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print(X.head(),"\n", y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=23)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model_Linear = svm.SVC(kernel='linear')
model_Linear.fit(X_train, y_train)
```

```python
y_pred_linear = model_Linear.predict(X_test)
```

```python
model_Linear.score(X_test, y_test)
```

```python
confusion_matrix(y_test, y_pred_linear)
```

```python
model_Poly = svm.SVC(kernel='poly')
model_Poly.fit(X_train, y_train)
```

```python
y_pred_poly = model_Poly.predict(X_test)
```

```python
model_Poly.score(X_test, y_test)
```

```python
confusion_matrix(y_test, y_pred_poly)
```

```python
model_Radial = svm.SVC(kernel='rbf')
model_Radial.fit(X_train, y_train)
```

```python
y_pred_radial = model_Radial.predict(X_test)
```

```python
model_Radial.score(X_test, y_test)
```

```python
confusion_matrix(y_test, y_pred_radial)
```
