```python
from sklearn.preprocessing import Binarizer, LabelEncoder, LabelBinarizer
import pandas as pd
```

```python
data = {
"City": ["New York", "Los Angeles", "Chicago", "Miami", "Boston"],
"Temperature (C)": [20, 25, 15, 30, 18],
"Humidity (%)": [65, 70, 60, 75, 55],
"Month": ["January", "February", "March", "April", "May"],
"Target": ["Yes", "No", "Yes", "No", "Yes"]
}

data = pd.DataFrame(data)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data['City'] = LabelEncoder().fit_transform(data['City'])
data['Month'] = LabelEncoder().fit_transform(data['Month'])

data['Target'] = LabelBinarizer().fit_transform(data['Target'])

bin1 = Binarizer(threshold=21.6)
data['Temperature (C)'] = bin1.fit_transform(data[['Temperature (C)']])

bin2 = Binarizer(threshold=65)
data['Humidity (%)'] = bin2.fit_transform(data[['Humidity (%)']])
data
```

```python
data.to_csv("Processed_Data.csv", index=False)
```
